```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# Store filepaths into variable
company_location = Path("raw data/Fortune_1000.csv")
company_2022 = Path("raw data/fortune_1000_2022.csv")
company_2021 = Path("raw data/fortune_1000_2021.csv")
```

```python
# Read in files.
company_location_df = pd.read_csv(company_location)
company_2022_df = pd.read_csv(company_2022)
company_2021_df = pd.read_csv(company_2021)
```

```python
# Display the companies with location dataset
company_location_df.head()
```

```python
# Display the companies of 2022
company_2022_df.head(10)
```

```python
# Display the companies of 2022
company_2021_df.head()
```

```python
# Merge company_2022 table with company_2021 table on name, with selected columns 
merged_df = company_2022_df.merge(company_2021_df, on="Name", suffixes=('', '_2021'))
```

```python
# Select the required columns from the merged dataframe
selected_columns = [
    "Name", "Revenue", "revenue(% change)", "profits in millions", "market value", "employees", 
    "Revenue_2021", "revenue(% change)_2021", "profits in millions_2021",
    "market value_2021", "employees_2021"
]
result_df = merged_df[selected_columns]
```

```python
result_df
```

```python
# Rename the columns in the resulting dataframe
result_df = result_df.rename(columns={
    "Revenue": "Revenue_2022",
    "revenue(% change)": "Revenue(% change)_2022",
    "profits in millions": "Profits In Millions_2022",
    "market value": "Market Value_2022",
    "employees": "Employees_2022"
})

# Capitalize all column names
result_df.columns = result_df.columns.str.capitalize()

# Print the dataframe with capitalized column names
result_df
```

```python
# Replace "-" with NaN in the dataframe
result_df.replace("-", np.nan, inplace=True)

# Check for null values in the dataframe
null_values = result_df.isnull()
```

```python
null_counts = result_df.isnull().sum()

# Print the count of null values
print(null_counts)
```

```python
# Remove rows with null values from the dataframe
result_df_cleaned = result_df.dropna()

# Print the cleaned dataframe
result_df_cleaned
```

```python
# Eliminate unwanted columns in the location dataset
company_location_df = company_location_df[["company", "sector", "city", "state", "ceo_woman"]]
```

```python
# Merging the selected columns from df1 with all columns from df2 based on "company" column
merged_df = result_df_cleaned.merge(company_location_df, left_on="Name", right_on="company")
merged_df.drop(columns=["company"], inplace=True)
merged_df
```

```python
# Capitalize all column names
merged_df.columns = merged_df.columns.str.capitalize()
merged_df
```

```python
# Replace "-" with NaN in the dataframe
merged_df.replace("-", np.nan, inplace=True)

# Check for null values in the dataframe
null_values = merged_df.isnull()
```

```python
null_counts = merged_df.isnull().sum()

# Print the count of null values
print(null_counts)
```

```python
# Define functions to convert strings to floats
def convert_dollar_string_to_float(dollar_string):
    return float(dollar_string.replace('$', '').replace(',', ''))

def convert_percent_string_to_float(percent_string):
    return float(percent_string.strip('%'))
```

```python
# Apply the conversion functions to relevant columns
dollar_columns = ['Revenue_2022', 'Profits in millions_2022', 'Market value_2022', 'Revenue_2021', 'Profits in millions_2021', 'Market value_2021']
percent_columns = ['Revenue(% change)_2022', 'Revenue(% change)_2021']

merged_df[dollar_columns] = merged_df[dollar_columns].applymap(convert_dollar_string_to_float)
merged_df[percent_columns] = merged_df[percent_columns].applymap(convert_percent_string_to_float)
```
