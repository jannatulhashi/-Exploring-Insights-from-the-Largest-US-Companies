```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# Store filepaths into variable
company_location = Path("raw data/Fortune_1000.csv")
company_2022 = Path("raw data/fortune_1000_2022.csv")
company_2021 = Path("raw data/fortune_1000_2021.csv")
```

```python
# Read in files.
company_location_df = pd.read_csv(company_location)
company_2022_df = pd.read_csv(company_2022)
company_2021_df = pd.read_csv(company_2021)
```

```python
# Display the companies with location dataset
company_location_df.head()
```

```python
# Display the companies of 2022
company_2022_df.head(10)
```

```python
# Display the companies of 2022
company_2021_df.head()
```

```python
# Merge company_2022 table with company_2021 table on name, with selected columns 
merged_df = company_2022_df.merge(company_2021_df, on="Name", suffixes=('', '_2021'))
```

```python
# Select the required columns from the merged dataframe
selected_columns = [
    "Name", "Revenue", "revenue(% change)", "profits in millions", "market value", "employees", 
    "Revenue_2021", "revenue(% change)_2021", "profits in millions_2021",
    "market value_2021", "employees_2021"
]
result_df = merged_df[selected_columns]
```

```python
result_df
```

```python

# Rename the columns in the resulting dataframe
result_df = result_df.rename(columns={
    "Name": "Company",
    "Revenue": "Revenue_2022",
    "revenue(% change)": "Revenue%change)_2022",
    "profits in millions": "Profits_in_millions_2022",
    "market value": "Market_value_2022",
    "employees": "Employees_2022"
})

# Modify the column names and capitalize them
result_df.columns = result_df.columns.str.replace(" ", "_").str.replace(r"\(|\)", "", regex=True).str.lower().str.capitalize()

# Print the dataframe with modified and capitalized column names
result_df





```

```python
# Replace "-" with NaN in the dataframe
result_df.replace("-", np.nan, inplace=True)

# Check for null values in the dataframe
null_values = result_df.isnull()
```

```python
null_counts = result_df.isnull().sum()

# Print the count of null values
print(null_counts)
```

```python
# Remove rows with null values from the dataframe
result_df_cleaned = result_df.dropna()

# Print the cleaned dataframe
result_df_cleaned
```

```python
result_df_cleaned.dtypes
```

```python
# Create a new DataFrame with the desired columns from the original DataFrame
result_df_cleaned_converted = result_df_cleaned.copy()

dollar_columns = ['Revenue_2022', 'Profits_in_millions_2022', 'Market_value_2022', 'Revenue_2021', 'Profits_in_millions_2021', 'Market_value_2021']
percent_columns = ['Revenue%change_2022', 'Revenue%change_2021']

# Apply the conversion functions directly to the new DataFrame
result_df_cleaned_converted[dollar_columns] = result_df_cleaned_converted[dollar_columns].applymap(lambda x: float(x.replace('$', '').replace(',', '')) if isinstance(x, str) else x)
result_df_cleaned_converted[percent_columns] = result_df_cleaned_converted[percent_columns].applymap(lambda x: float(x.strip('%')) if isinstance(x, str) else x)
result_df_cleaned_converted["Employees_2022"] = result_df_cleaned_converted["Employees_2022"].str.replace(',', '').astype("int64")

# Assign the new DataFrame back to the original DataFrame
result_df_cleaned = result_df_cleaned_converted


```

```python
result_df_cleaned.dtypes
```

```python
# Select columns for 2022 data
columns_2022 = ['Company', 'Revenue_2022', 'Revenue%change_2022', 'Profits_in_millions_2022', 'Market_value_2022', 'Employees_2022']

# Create separate DataFrames for 2022 data and 2021 data
data_2022 = result_df_cleaned[columns_2022]

# Print the first few rows of the 2022 data DataFrame
print("2022 Data:")
data_2022.head(10)

```

```python
#Select columns for 2021 data
columns_2021 = ['Company', 'Revenue_2021', 'Revenue%change_2021', 'Profits_in_millions_2021', 'Market_value_2021', 'Employees_2021']

# Create separate DataFrames for 2022 data and 2021 data
data_2021 = result_df_cleaned[columns_2021]

# Print the first few rows of the 2021 data DataFrame
print("\n2021 Data:")
data_2021.head(10)
```

```python
# Export data_2021 and data_2022 as CSV files.

data_2021.to_csv("Resources/company_data_2021.csv", index=False)

data_2022.to_csv("Resources/company_data_2022.csv", index=False)
```
